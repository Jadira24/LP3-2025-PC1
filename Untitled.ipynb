DETECCIÓN DE CÁNCER DE CUELLO UTERINO (CA_CERVIX)

IMPORTAR LIBRERÍAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

CARGAR EL DATASET

```python
dataframe = pd.read_csv('DSCancerDeCuelloUterinoDatosPerdidos.csv', sep=';')
dataframe.head(10)
```

DEFINIENDO LAS VARIABLES

```python
#Variable dependiente: cancer de cuello uterino
y = dataframe['CA_CERVIX']

#Variable independiente: Todos los atributos de cada columna
x = dataframe.drop(columns=['CA_CERVIX'])
```

DISTRIBUIR LA VARIABLE RESULTADO

```python
sb.catplot(x='CA_CERVIX', data=dataframe, kind='count', height=4, aspect=2)
plt.title('Distribucion de Cancer de Cuello Uterino (0=No, 1=Si)')
plt.xticks([0,1],['No Cancer', 'Si Cancer'])
plt.show()
```

SEPARAR EN DATOS DE ENTRENAMIENTO Y PRUEBA

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=30)
```

MODELO KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train, y_train)
```
